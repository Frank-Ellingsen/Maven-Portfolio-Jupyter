```python
import  numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("powerconsumption.csv",usecols=['Datetime','PowerConsumption_Zone1','PowerConsumption_Zone2','PowerConsumption_Zone3'],parse_dates=['Datetime'])
df['Total PowerConsumption']=df['PowerConsumption_Zone1']+df['PowerConsumption_Zone2']+df['PowerConsumption_Zone3']
df['Hour']=df['Datetime'].dt.hour
df['DayofWeek']=df['Datetime'].dt.dayofweek
df
```

```python
df.set_index('Datetime',inplace=True)
df
```

```python
dfr=df.resample("H").mean()
dfr
```

```python
dfrj=dfr.loc["2017-01"]
dfr
```

```python

dfrj[['PowerConsumption_Zone1','PowerConsumption_Zone2','PowerConsumption_Zone3']].plot.area(
    title="Hourly Comsumption Jan 2017",xlabel="Date",ylabel="KwH")
sns.despine()
plt.show()
```

```python
table=pd.pivot_table(df,index='Hour',columns='DayofWeek',values='Total PowerConsumption',
                     aggfunc={'Total PowerConsumption':np.mean})

sns.heatmap(table,cmap='coolwarm').set(title='Power Consumption by Hour & Day of Week');
```
